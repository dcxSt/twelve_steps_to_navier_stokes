```python
import numpy as np
from matplotlib import pyplot as plt
import time, sys
```

```python
%matplotlib inline
```

```python
nx = 41 
dx = 2 / (nx-1)
nt = 25 # no of timesteps we want to calculate
dt = 0.025
c = 1 # wavespeed of c=1
```

```python
# Initial conditions
# u = np.ones(nx)
u = np.exp(-np.linspace(-4,4,nx)**2)
# set u=2 between 0.5 and 1
# u[int(.5 / dx):int(1 / dx + 1)] = 2
plt.plot(np.linspace(0,2,nx),u)
```

```python
plt.figure(figsize=(15,8))
un = np.ones(nx)
for n in range(nt):
    un = u.copy()
    for i in range(1,nx):
        u[i] = un[i] * (1 -  dt / dx * (un[i] - un[i-1]))
    plt.plot(np.linspace(0,2,nx),u)
```
